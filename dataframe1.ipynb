```python

import pandas as pd

df = pd.read_csv("world_population.csv")
df.columns = df.columns.str.strip()

df_melted = df.melt(
    id_vars=['Country/Territory', 'Continent'],
    value_vars=[
        '1970 Population', '1980 Population', '1990 Population',
        '2000 Population', '2010 Population', '2015 Population',
        '2020 Population', '2022 Population'
    ],
    var_name='Year',
    value_name='Population'
)


df_melted['Year'] = df_melted['Year'].str.extract(r'(\d{4})').astype(int)


print(df_melted.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("world_population.csv")

df['Continent'] = pd.to_numeric(df['Continent'], errors='coerce')
df['2022 Population'] = pd.to_numeric(df['2022 Population'], errors='coerce')
df['2020 Population'] = pd.to_numeric(df['2020 Population'], errors='coerce')

median_age_column5 = df['Continent'].median()
median_age_column6 = df['2022 Population'].median()
median_age_column7 = df['2020 Population'].median()

df['Continent'].fillna(median_age_column5, inplace=True)
df['2022 Population'].fillna(median_age_column6, inplace=True)
df['2020 Population'].fillna(median_age_column7, inplace=True)

age_categories = ['0-14', '15-64', '65+']

df_stacked = df[['Continent', '2022 Population', '2020 Population']].copy()
df_stacked.columns = age_categories


total_population = df_stacked.sum()


plt.figure(figsize=(10, 6))
plt.bar(age_categories, total_population)
plt.xlabel('Age Category')
plt.ylabel('Population Count')
plt.title('Bar Chart for Age Distribution in the Population (2022)')
plt.show()
```
